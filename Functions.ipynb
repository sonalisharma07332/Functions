1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python

list_input=input("Enter the List of numbers.")
list1=[int(x) for x in list_input.split(',')]
print(list1)
def numbers(list1):
    sum_even=0
    for i in list1:
        if i%2==0:
            sum_even=sum_even+i
    return sum_even    
print("Sum of even numbers:", numbers(list1))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
num = int(input("Enter a number: "))
def prime(num):
    if num < 2 or num > 200:
        print("Number must be between 2 and 200.")
        return
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print(num, "is not a prime number.")
            return
    print(num, "is a prime number.")
prime(num)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.


```python
class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms  # Total number of terms to generate
        self.count = 0              # Counter to keep track of terms generated
        self.a = 0                  # First number in Fibonacci sequence
        self.b = 1                  # Second number

    def __iter__(self):
        return self  # Returns the iterator object itself

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration  # Stop iteration when limit is reached

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b
fib = Fibonacci(10)  # Create an instance to generate first 10 Fibonacci numbers
for num in fib:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


```python
def powers_of_two(max_exponent):
    """
    Yields powers of 2 from 0 up to max_exponent (inclusive).
    """
    for exponent in range(max_exponent + 1):
        yield 3 ** exponent
for power in powers_of_two(10):
    print(power)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.


```python
def read_file_line_by_line(file_path):
    """
    Generator that reads a file line by line and yields each line.
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            yield line.rstrip('\n')  # Remove newline

# Example usage:
for line in read_file_line_by_line(r'H:\My Drive\PW_COURSE\functions\myself.txt'):
    print(line)
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.


```python
# Step 1: Take user input for number of tuples
n = int(input("Enter the number of tuples: "))

# Step 2: Input each tuple
tuple_list = []
for i in range(n):
    a = input(f"Enter two values for tuple {i+1} (separated by space): ").split()
    if len(a) != 2:
        print("Invalid input. Please enter exactly two values.")
        exit()
    tuple_list.append((int(a[0]), int(a[1])))

# Step 3: Sort using lambda based on second element
sorted_list = sorted(tuple_list, key=lambda x: x[1])

# Step 4: Print the result
print("\nSorted list based on second element of each tuple:")
for t in sorted_list:
    print(t)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


```python
# Step 1: Take user input for temperatures in Celsius
celsius = input("Enter temperatures in Celsius (separated by spaces): ").split()

# Step 2: Convert input strings to floats
celsius = list(map(float, celsius))

# Step 3: Use map() to convert to Fahrenheit
fahrenheit = list(map(lambda c: (9/5) * c + 32, celsius))

# Step 4: Print the results
print("\nTemperatures in Fahrenheit:")
for c, f in zip(celsius, fahrenheit):
    print(f"{c}°C = {f}°F")
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.



```python
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    # Filter out characters that are vowels
    filtered_chars = filter(lambda char: char not in vowels, s)
    # Join the filtered characters back into a string
    return ''.join(filtered_chars)

# Take input from user
input_str = input("Enter a string: ")

# Remove vowels
result = remove_vowels(input_str)

print("String without vowels:", result)
```

11.Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.



```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

def process_orders(order_list):
    result = []
    for order in order_list:
        order_number = order[0]
        quantity = order[2]
        price_per_item = order[3]
        total = quantity * price_per_item
        if total < 100:
            total += 10  # surcharge
        result.append((order_number, round(total, 2)))  # rounded to 2 decimals
    return result

# Run the function and print results
processed = process_orders(orders)
print(processed)
```

Write a Python program using lambda and map.



```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

# Lambda function to calculate total with surcharge
calc_total = lambda order: (
    order[0], 
    round(order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0), 2)
)

# Use map to apply the lambda to each order
result = list(map(calc_total, orders))

print(result)
```
